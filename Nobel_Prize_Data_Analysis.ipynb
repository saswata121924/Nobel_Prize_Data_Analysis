```python
import pandas as pd
df_data = pd.read_csv("nobel_prize_data.csv")
print(df_data.shape)
df_data.head()
```

```python
df_data.info()
```

```python
df_data.describe()
```

```python
df_data["birth_date"] = pd.to_datetime(df_data["birth_date"])
df_data["birth_date"].info()
```

```python
df_data.isna().sum()
```

```python
df_data[df_data.isna().values]
```

```python
share_values = df_data["prize_share"].str.split('/', expand=True)
numerator = pd.to_numeric(share_values[0])
denominator = pd.to_numeric(share_values[1])
df_data["share_pct"] = numerator/denominator
df_data.head()
```

```python
gender_ratio = df_data.groupby(by='sex')["full_name"].count()
gender_ratio.head()
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
```

```python
px.pie(gender_ratio,
       names=gender_ratio.index,
       values=gender_ratio.values,
       title="Percentage of Male vs Female winners",
       hole=0.4,
       labels=gender_ratio.index)
```

```python
print("The first 3 female Nobel Laureates are:")
df_data[df_data["sex"]=="Female"].sort_values(by="year", ascending=True).head(3)
```

```python
is_winner = df_data.duplicated(subset=["full_name"], keep=False)
multiple_winners = df_data[is_winner]
print(f"There are {multiple_winners.full_name.nunique()} who were awarded the Nobel Prize more than once")
multiple_winners[["year", "category", "laureate_type", "full_name"]]
```

```python
category_wise_data = df_data.groupby(by="category").count()["full_name"]
category_wise_data.head()
```

```python
px.bar(category_wise_data,
       x=category_wise_data.index,
       y=category_wise_data.values,
       color=category_wise_data.values,
       labels=category_wise_data.index,
       color_continuous_scale="Aggrnyl",
       title="Number of Prizes Awarded per Category"
       ).update_layout(xaxis_title="Categories",
                       yaxis_title="Number of prizes",
                       coloraxis_showscale= False)
```

```python
economics_data = df_data[df_data['category']=="Economics"].sort_values(by="year")
economics_data.head()
```

```python
category_wise_data = df_data.groupby(["category", "sex"], as_index=False).agg({'prize': pd.Series.count})
category_wise_data.sort_values(by="prize", ascending=False, inplace=True)
category_wise_data
```

```python
px.bar(category_wise_data,
       x=category_wise_data.category,
       y=category_wise_data.prize,
       color=category_wise_data.sex,
       title="Number of Prizes Awarded per Category split by Men and Women"
       ).update_layout(xaxis_title="Nobel Prize Category",
                       yaxis_title="Number of prizes won"
                       )
```

```python
number_of_prizes_per_year = df_data.groupby("year").count()["prize"]
rolling_window_for_prizes = number_of_prizes_per_year.rolling(window=5).mean()
rolling_window_for_prizes
```

```python
plt.figure(figsize=(16,8), dpi=200)
plt.title("Number of Nobel Prize awarded each year", fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45
          )
plt.xlim(1900, 2021)
plt.scatter(x=number_of_prizes_per_year.index,
            y=number_of_prizes_per_year.values,
            c="dodgerblue",
            alpha=0.5,
            s=80
            )
plt.plot(rolling_window_for_prizes,
         c="crimson",
         linewidth=2)
```

```python
average_prize_share = df_data.groupby("year").agg({"share_pct": pd.Series.mean})
rolling_window_for_prize_share = average_prize_share.rolling(window=5).mean()
```

```python
plt.figure(figsize=(16,8), dpi=200)
plt.title("Prize Share vs year", fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(ticks=np.arange(1900, 2021, step=5),
           fontsize=14,
           rotation=45
          )
ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.invert_yaxis()
ax1.set_xlim(1900, 2021)
ax1.scatter(x=number_of_prizes_per_year.index,
            y=number_of_prizes_per_year.values,
            c="dodgerblue",
            alpha=0.5,
            s=80
            )
ax1.plot(rolling_window_for_prizes,
         c="crimson",
         linewidth=2)
ax2.plot(rolling_window_for_prize_share,
         c="grey",
         linewidth=3
         )
```

```python
df_countries = df_data.groupby(by=["birth_country_current", "ISO"], as_index=False).agg({"prize": pd.Series.count})
df_countries.sort_values("prize", ascending=False, inplace=True)
px.choropleth(df_countries,
              locations= "ISO",
              color="prize",
              hover_name="birth_country_current",
              color_continuous_scale = px.colors.sequential.matter,
              title="Countries vs Number of Nobel prizes received"
              )
```

```python
category_prizes = df_data.groupby(by=["birth_country_current", "category"], as_index=False).count()
category_prizes = category_prizes.sort_values(by="prize", ascending=False)[["birth_country_current","category","prize"]]

top_20_countries = df_data.groupby(by="birth_country_current", as_index=False).count()[["birth_country_current","prize"]]
top_20_countries = top_20_countries.sort_values("prize")[-20:]

df_merge = pd.merge(category_prizes, top_20_countries, on="birth_country_current")
df_merge.columns = ["birth_country_current", "category", "cat_prize", "total_prize"]
df_merge.sort_values(by="total_prize", inplace=True)
```

```python
px.bar(df_merge,
       x="cat_prize",
       y="birth_country_current",
       title="Top 20 Countries by Number of prizes and Category",
       color="category",
       orientation = "h",
       ).update_layout(xaxis_title="Number of Prizes",
                       yaxis_title="Country")
```

```python
year_data = df_data.groupby(by=["birth_country_current","year"], as_index=False).count()[["year", "birth_country_current", "prize"]]
year_data.sort_values(by="year", inplace=True)
year_cumulative_data = year_data.groupby(by=["birth_country_current","year"]).sum().groupby(level=[0]).cumsum()
year_cumulative_data.reset_index(inplace=True)
year_cumulative_data
```

```python
px.line(year_cumulative_data,
        x="year",
        y="prize",
        color="birth_country_current",
        hover_name="birth_country_current",
        title="Number of prizes over the years"
        ).update_layout(xaxis_title="Year",
                        yaxis_title="Number of Prize"
                        )
```

```python
organization_data = df_data.groupby(by="organization_name", as_index=False).count()
organization_data.sort_values("prize", inplace=True)
organization_data = organization_data[["organization_name", "prize"]][-20:]
```

```python
px.bar(organization_data,
       x="prize",
       y="organization_name",
       color="prize",
       title="Top 20 Research Institutions by Number of Prizes",
       ).update_layout(xaxis_title="Number of Prizes",
                       yaxis_title="Organization Names",
                       coloraxis_showscale=False)
```

```python
organization_data = df_data.groupby(by="organization_city", as_index=False).count()
organization_data.sort_values("prize", inplace=True)
organization_data = organization_data[["organization_city", "prize"]][-20:]
px.bar(organization_data,
       x="prize",
       y="organization_city",
       color="prize",
       title="Cities with the most Researchers",
       color_continuous_scale = px.colors.sequential.haline
       ).update_layout(xaxis_title="Number of Prizes",
                       yaxis_title="Organization Cities",
                       coloraxis_showscale=False)
```

```python
birth_city_data = df_data.groupby(by="birth_city", as_index=False).count()
birth_city_data.sort_values("prize", inplace=True)
birth_city_data = birth_city_data[["birth_city", "prize"]][-20:]
px.bar(birth_city_data,
       x="prize",
       y="birth_city",
       color="prize",
       title="Birth Place of most Researchers",
       ).update_layout(xaxis_title="Number of Prizes",
                       yaxis_title="Birth Cities",
                       coloraxis_showscale=False)
```

```python
df_organization = df_data.groupby(by=["organization_country","organization_city","organization_name"], as_index=False).agg({"prize": pd.Series.count})
df_organization = df_organization.sort_values(by="prize", ascending=False)
```

```python
px.sunburst(df_organization,
            path=["organization_country","organization_city","organization_name"],
            values="prize",
            title="Where do Discoveries take place?"
            ).update_layout(xaxis_title="Number of prizes",
                            yaxis_title="City",
                            )
```

```python
df_data["winning_age"] = (df_data["year"] - df_data["birth_date"].dt.year)
display(df_data.head())
```

```python
display(df_data.nlargest(n=1, columns="winning_age"))
display(df_data.nsmallest(n=1, columns="winning_age"))
```

```python
import seaborn as sns
```

```python
df_data.winning_age.describe()
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.histplot(df_data,
             x="winning_age")
plt.xlabel("Age")
plt.title("Distribution of Age on Receipt of prize")
```

```python
plt.figure(figsize=(8,4), dpi=200)
plt.title("Regression plot for Winning Age vs Year")
plt.xlim(1900, 2020)
with sns.axes_style("darkgrid"):
  sns.regplot(df_data,
              x="year",
              y="winning_age",
              lowess = True,
              scatter_kws = {"alpha": 0.4},
              line_kws={"color": "black"}
              )
plt.xlabel("Year")
plt.ylabel("Winning Age")
```

```python
px.box(df_data,
       x="category",
       y="winning_age",
       color="category",
       title="Winning Age per Category"
       ).update_layout(xaxis_title="Category",
                       yaxis_title="Winning Age")
```

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
  sns.lmplot(df_data,
            x="year",
            y="winning_age",
            hue="category",
            row="category",
            aspect=2,
            lowess=True,
            scatter_kws={"alpha":0.4},
            line_kws={"color": "black"}
            )
```
